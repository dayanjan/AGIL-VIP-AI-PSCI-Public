## ðŸ“š Background

### What are Molecular Fingerprints?

**Molecular fingerprints** are binary or count vectors that encode the structural features of a molecule. Think of them as a molecule's "barcode" â€” a standardized representation that allows computers to compare molecules quickly.

Fingerprints are foundational to:
- **Virtual screening**: Finding compounds similar to known actives
- **Clustering**: Grouping molecules by structural similarity
- **Machine learning**: Input features for activity prediction models
- **Database searching**: Efficient similarity searches across millions of compounds

### Types of Fingerprints

**1. Structural Keys (MACCS Keys)**
- 166 predefined structural features (aromatic rings, specific functional groups)
- Each bit = presence/absence of a specific substructure
- Fast to compute, interpretable
- Limited by predefined features

**2. Circular Fingerprints (Morgan/ECFP)**
- Generated by examining atom neighborhoods up to a specified radius
- ECFP4 = radius 2 (diameter 4 bonds), ECFP6 = radius 3
- More flexible and comprehensive than MACCS
- Gold standard for similarity searching

### The Tanimoto Coefficient

The **Tanimoto coefficient** (also called Jaccard index) measures similarity between two fingerprints:

$$T(A, B) = \frac{|A \cap B|}{|A \cup B|} = \frac{c}{a + b - c}$$

Where:
- $a$ = bits "on" in fingerprint A
- $b$ = bits "on" in fingerprint B
- $c$ = bits "on" in both A and B

Interpretation:
- **T = 1.0**: Identical fingerprints (very similar structures)
- **T > 0.7**: Generally considered "similar" for activity
- **T > 0.85**: Very high similarity, likely same scaffold
- **T < 0.4**: Dissimilar molecules

### Key Concepts

- **Fingerprint radius**: How many bonds from each atom to consider (Morgan)
- **Fingerprint length**: Number of bits in the vector (typically 1024 or 2048)
- **Bit collision**: Different features mapping to the same bit (can reduce accuracy)
- **Similarity matrix**: Pairwise similarities for all compounds in a set
