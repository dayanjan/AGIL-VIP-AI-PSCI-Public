## Background

### The IMRAD Structure

Scientific papers follow the IMRAD format:
- **I**ntroduction: Why did you do this?
- **M**ethods: How did you do it?
- **R**esults: What did you find?
- **A**nd **D**iscussion: What does it mean?

### Figure Design Principles

1. **Clarity**: Every element should serve a purpose
2. **Accuracy**: Data representation must be truthful
3. **Accessibility**: Consider colorblind viewers
4. **Consistency**: Use same style across all figures

### The Pyramid Principle

Start with the conclusion, then provide supporting details:

```
         [Main Finding]
        /      |      \
   [Support] [Support] [Support]
      |         |         |
   [Data]    [Data]    [Data]
```

### Plain Language Guidelines (NIH)

- Use common words instead of jargon
- Keep sentences short (15-20 words)
- Define technical terms when necessary
- Use active voice
